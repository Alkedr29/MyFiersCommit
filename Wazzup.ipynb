## 1. Метрики оценки онбординга:

Задача ондбординга привести зарегистрированного пользователя к использованию продукта.(активация)
Если % активированных, среди тех кто просмотрел онбординг выше чем среди тех кто его скипнул - онбординг приносит пользу.

Способы оценки:
* Доля пользователей полностью прошедших анбординг. Минус данной оценки: мы оцениваем только комфортность прохождения, но не качество информации. Используем только в паре с другими, например с долей активации.
* Сравниваем % активаций. Основная цель онбординга, показать, что наш продукт это то, что ему нужно. Если клиент активирован, значит цель достигнута.
* Общая прибыль принесенная от среднего клиента. Клиент не только попробовал, но и продлил подписку, значит получил всю необходимую информацию о пользовании для себя.Учитывая разнообразие подписок: чем дороже, тем более продвинутые возможности, а значит такая информация была точно донесена до клиента. Минус данной метрики: в случаях, когда пользователь никак не нуждается в расширении возможностей. Не лишним будет узнать на сколько пользователь осведомлен в расширении возможностей(опрос, пробный период, спец. предложения).
* Средняя оценка нашего сервиса от клиентов. Если клиент доволен, значит продукт ему полезен и покрывает его потребности, следовательно, информация точно донесена. Минусом будет множество факторов связанных скорее с качеством самого продукта, чем с информацией о пользовании. Данная оценка будет полезна если пользователь недавно с нами, и практически бесполезна если что-то в продукте были изменения.
* % оттока клиентов. Отнесем отток на последнее место т.к. есть огромное множество факторов по которым клиент мог отказаться: личные обстоятельства, качество продукта, стоимость, конкуренция и многие другие. Данная метрика сработает в том множестве клиентов, для которых пользование сервисом было сложным в ориентировании.

Чтобы проверить заменим видео на текстовые и замерим % прошедших анбординг. Эксперимент удачен если % прошедших онбординг выше на 5% и % активаций не ниже нынешнего.


## 3. Продуктовая команда выяснила, что в последние месяцы конверсия из тестирования продукта на бесплатном пробном периоде в покупку подписки сильно упала:

#### Оглянемся вокруг и посмотрим, что изменилось, обратимся к данным, где это возможно. На основе данных построим и будем проверять гипотезы. 
Падение конверсии от тестирования к активации означает что клиенты приходят и по какой-то причине реже приобретают подписку. Если есть информация после каких изменений произошло падение обратимся сразу к ней, если такой информации нет, обратимся к данным:
    Группируем клиентов которые по факту активации, сравниваем кто они, и чем они глобально отличаются между собой. Целевой признак активация, проведем корреляцию целевого признака с прочими. В первую очередь рассматриваем подробно сильно коррелирующие признаки, далее просмотрим такую же корреляцию за прошлый период и рассмотрим признаки корреляция с которыми изменилась. Таким образом мы имеет хороший шанс уменьшить радиус поиска. Например:
    
* Каналы привлечения клиента. Проведем исследование по каналам привлечения, сравним конверсию до и после поканально, сравним не только доли, но и суммарное количество на период. Возможно пользователей в целом стало больше из-за некоторого канала где не совсем корректна предоставлена информация и они оказались в сервисе ошибочно. В таком случае среднее количество активаций будет в порядке, а конверсия искусственно занижена.
-- H-0: Я верю, что если исправить проблему на аномальном канале привлечения, конверсия тестирования в активацию нормализуется. Эксперимент считается удачным, если конверсия нормализовалась.


* Пример с платежными системами. Проверим способы оплаты, убедимся что никакой сервис не отвалился и все предлагаемые способы корректно работают. Если нашли такой сервис и проблему с платежной системой можно исправить, строим гипотезу:
-- H-0: Я верю, что если вернуть возможность оплаты через данную систему, конверсия тестирования в активацию нормализуется. Эксперимент считается удачным, если конверсия нормализовалась.


* Посчитаем конверсию по самостоятельно. Допустим находим наш разнящийся признак, в который данные по какой-то причине были записаны в другом формате. Строим гипотезу: 
-- H-0: Я верю, что если привести стандартный формат данных в признаке, то при верном расчете конверсия тестирования в активацию нормализуется. Эксперимент считается удачным, если конверсия нормализовалась.


* Визуально проверим платформу, обратим внимание, что поменялось, дизайн, интерфейс, контент и способ подачи контента в самом онбординге. Например, заметили большую красную кнопку которой раньше не было. Предположим это негативно влияет на пользователей и строим гипотезу: 
-- Н-0 Я верю, что если убрать\заменить большую красную кнопку, конверсия тестирования в активацию нормализуется. Эксперимент считается удачным, если конверсия нормализовалась.


* Изменили длительность тест-периода. Предположим введённый тест-период увеличили и клиенты не оплачивали подписку потому, что на момент сбора данных для расчета конверсии он еще действовал. Строим гипотезу: 
-- H-0 Я верю, что если ничего не менять, конверсия тестирования в активацию нормализуется. Эксперимент считается удачным если конверсия нормализовалась.
