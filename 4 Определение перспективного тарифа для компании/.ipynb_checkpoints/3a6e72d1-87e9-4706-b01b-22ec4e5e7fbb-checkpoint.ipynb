# Определение перспективного тарифа для телеком-компании

Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

Нам предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

Посчитаем:

  * количество сделанных звонков и израсходованных минут разговора по месяцам

  * количество отправленных сообщений по месяцам

  * объем израсходованного интернет-трафика по месяцам

  * помесячную выручку с каждого пользователя

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from scipy import stats as st
```

```python
calls = pd.read_csv('calls.csv')
internet = pd.read_csv('internet.csv')
messages = pd.read_csv('messages.csv')
tariffs = pd.read_csv('tariffs.csv')
users = pd.read_csv('users.csv')
```

 ## 1 Подготовка данных

```python
display(calls.info())
calls.head()
```

<div class="alert alert-info" role="alert">
Исключим звонки с 0 длительностью из анализа,т.к. в рассчетах звонков они не участвуют.
</div>

```python
calls = calls.query('duration > 0')
```

```python
calls['call_date'] = calls['call_date'].astype('datetime64')
calls['duration'] = np.ceil(calls['duration']) 
#calls['month'] = calls['call_date'].dt.month
calls['month'] = calls['call_date'].astype('datetime64[M]')
calls
```

```python
calls
```

<div class="alert alert-info" role="alert">
Посчитаем количество и длительность звонков,интернет трафик и отправленые смс  по мясяцам у каждого клиента.
</div>

```python
calls_user = calls.groupby(['user_id','month'])['duration'].agg(['sum','count'])
```

```python
calls_user = calls_user.reset_index()
calls_user.columns =  ['user_id','month','minuts','calls']
calls_user
```

```python
calls_user.duplicated().sum()
```

<div class="alert alert-info" role="alert">
Исключим из анализа месяца когда интернетом не пользовались и уберем дублирующий индексы столбец.
</div>

```python
internet = internet.drop(['Unnamed: 0'], axis=1)
internet['session_date'] = internet['session_date'].astype('datetime64')
```

```python
#internet['month'] = internet['session_date'].dt.month
internet['month'] = internet['session_date'].astype('datetime64[M]')
internet = internet.query('mb_used > 0')
internet_month_user = internet.groupby(['user_id','month'])['mb_used'].sum().reset_index()
internet_month_user['gb_used'] = np.ceil(internet_month_user['mb_used'] / 1024)
internet_month_user
```

```python
messages
```

```python
messages.info()
```

```python
messages['message_date'] = messages['message_date'].astype('datetime64')
messages['month'] = messages['message_date'].astype('datetime64[M]')
message_user_month = messages.groupby(['user_id','month'])['id'].count().reset_index()
message_user_month.columns = ['user_id', 'month', 'messages']
message_user_month
```

<div class="alert alert-info" role="alert">
Обьединим таблицы полученые таблицы.
</div>

```python
minutes_month = calls_user[['user_id','month','minuts']]
internet_month = internet_month_user[['user_id', 'month', 'gb_used']]
message_month = message_user_month
tariff_month = users[['user_id','tariff','city']]
```

```python
general = minutes_month.merge(internet_month,on=['user_id','month'],how='outer')
general = general.merge(message_month,on=['user_id','month'],how='outer')
general = general.merge(tariff_month,on='user_id')
general
```

```python
general.info()
```

<div class="alert alert-info" role="alert">
Пропуски появились в результате того,что некоторые пользовались не всеми услугами,заполним 0.
</div>

```python
general = general.fillna(0)
general[['minuts', 'messages']] = general[['minuts', 'messages']].astype('int')
general
```

<div class="alert alert-info" role="alert">
обьединим таблицу с тарифами с основной таблицей.
</div>

```python
tariffs = tariffs.rename(columns={'tariff_name' : 'tariff'})
general = general.merge(tariffs,on='tariff',how='left')
general
```

<div class="alert alert-info" role="alert">
Посчитаем доход от каждого клиента по месяцам.
</div>

```python
def income(row):
    pay = row['rub_monthly_fee']
    if row['minuts'] > row['minutes_included']:
            pay += ((row['minuts'] - row['minutes_included']) * row['rub_per_minute'])
    if row['gb_used'] > row['gb_per_month_included']:
            pay += ((row['gb_used'] - row['gb_per_month_included']) * (row['rub_per_gb']))
    if row['messages'] > row['messages_included']:
            pay += ((row['messages'] - row['messages_included']) * row['rub_per_message'])
    return pay
general['pay'] = general.apply(income, axis=1)
general.head(20)
```

 ## 2 Анализ данных

<div class="alert alert-info" role="alert">
Построим гистограмы и посмотри распределение как клиенты используют лимит по звонкам,интернету и сообщениям в разных тарифах. 
</div>

```python
gen_smart = general.query('tariff == "smart" ')
gen_ultra = general.query('tariff == "ultra" ')
```

```python
f,axs = plt.subplots(1,2,figsize=(16,7))
axs[0].hist(gen_smart['minuts'],bins=(50))
axs[0].grid()
axs[0].set_title('smart')
axs[0].set_xlabel('Minuts')
axs[0].set_ylabel('Count')
axs[1].hist(gen_ultra['minuts'],bins=(50))
axs[1].grid()
axs[1].set_title('ultra')
axs[1].set_xlabel('Minuts')
axs[1].set_ylabel('Count');
```

```python
smart_minutes = gen_smart['tariff'].reset_index()
smart_minutes['mean'] = int(gen_smart['minuts'].mean())
smart_minutes['median'] = int(gen_smart['minuts'].median())
smart_minutes['var'] = int(np.var(gen_smart['minuts']))
smart_minutes['std'] = int(np.std(gen_smart['minuts']))

ultra_minutes = gen_ultra['tariff'].reset_index()
ultra_minutes['mean'] = int(gen_ultra['minuts'].mean())
ultra_minutes['median'] = int(gen_ultra['minuts'].median())
ultra_minutes['var'] = int(np.var(gen_ultra['minuts'],ddof=0))
ultra_minutes['std'] = int(np.std(gen_ultra['minuts']))

gen_minutes = smart_minutes.head(1).merge(ultra_minutes.head(1),how='outer')
gen_minutes
```

<div class="alert alert-info" role="alert">
В основном на смарте,клиентам требуется ~400-450мин.    

На ультре лимит в 3000 мин никогда не расходуется,в редких случаях тратят 1700мин.В основном 550 минут клиентам хватает.Значительная часть даже не пользуется звонками,возможно тариф им интересен только из-за интернета.
</div>

```python
f,axs = plt.subplots(1,2,figsize=(16,7))
axs[0].hist(gen_smart['gb_used'],bins=(50));
axs[0].grid()
axs[0].set_title('smart')
axs[0].set_xlabel('Gb used')
axs[0].set_ylabel('Count')
axs[1].hist(gen_ultra['gb_used'],bins=(50));
axs[1].grid()
axs[1].set_title('ultra')
axs[1].set_xlabel('Gb used')
axs[1].set_ylabel('Count');
```

```python
smart_internet = gen_smart['tariff'].reset_index()
smart_internet['mean'] = int(gen_smart['gb_used'].mean())
smart_internet['median'] = int(gen_smart['gb_used'].median())
smart_internet['var'] = int(np.var(gen_smart['gb_used']))
smart_internet['std'] = int(np.std(gen_smart['gb_used']))

ultra_internet = gen_ultra['tariff'].reset_index()
ultra_internet['mean'] = int(gen_ultra['gb_used'].mean())
ultra_internet['median'] = int(gen_ultra['gb_used'].median())
ultra_internet['var'] = int(np.var(gen_ultra['gb_used']))
ultra_internet['std'] = int(np.std(gen_ultra['gb_used']))


gen_internet = smart_internet.head(1).merge(ultra_internet.head(1),how='outer')
gen_internet = gen_internet.drop(['index'],axis=1)
gen_internet
```

<div class="alert alert-info" role="alert">
В основном на смарте интернет расходуют полностью.Чаще больше чем меньше.Можно сказать что данного лимита немного нехватает.

На ультре в большенстве случаев клиентам хватает 20Гб на месяц,за предел лимита вылазиет малая часть.
</div>

```python
f,axs = plt.subplots(1,2,figsize=(16,7))
axs[0].hist(gen_smart['messages'],bins=(50))
axs[0].grid()
axs[0].set_title('smart')
axs[0].set_xlabel('Messages')
axs[0].set_ylabel('Count')
axs[1].hist(gen_ultra['messages'],bins=(50))
axs[1].grid()
axs[1].set_title('ultra')
axs[1].set_xlabel('Messages')
axs[1].set_ylabel('Count');
```

```python
smart_message = gen_smart['tariff'].reset_index()
smart_message['mean'] = int(gen_smart['messages'].mean())
smart_message['median'] = int(gen_smart['messages'].median())
smart_message['var'] = int(np.var(gen_smart['messages']))
smart_message['std'] = int(np.std(gen_smart['messages']))
ultra_message = gen_ultra['tariff'].reset_index()
ultra_message['mean'] = int(gen_ultra['messages'].mean())
ultra_message['median'] = int(gen_ultra['messages'].median())
ultra_message['var'] = int(np.var(gen_ultra['messages'],ddof=0))
ultra_message['std'] = int(np.std(gen_ultra['messages']))

gen_message = smart_message.head(1).merge(ultra_message.head(1),how='outer')
gen_message = gen_message.drop(['index'],axis=1)
gen_message
```

<div class="alert alert-info" role="alert">
Дополнительно платят за смс только на смарте,в основном для всех тарифов хватает 50 смс.
    
На ультре даже в редчайших случаях не расходуют и 1/4 от бесплатного лимита.
    
Многие клиенты вообще не пользуются смс независимо от тарифа.

 ## 3 Проверка гипотез

<div class="alert alert-info" role="alert">
    Проверим гипозу:
    
h0: средняя выручка с пользователей тарифа «Ультра» равна средней выручке с тарифа «Смарт»
    
h1: средняя выручка с пользователей тарифа «Ультра» не равна средней выручке с тарифа «Смарт»
</div>

```python
general_smart = general.query('tariff == "smart"')
general_ultra = general.query('tariff == "ultra"')
alpha = .05
results = st.ttest_ind(general_smart['pay'],general_ultra['pay'])
display(results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не отвергаем нулевую гипозезу')
```

<div class="alert alert-info" role="alert">
Получается средняя выручка на разных тарифах разная.
</div>

<div class="alert alert-info" role="alert">
Проверим гипозу:
    
h0: Cредняя выручка пользователей из Москвы равна выручке пользователей из других регионов.
    
h1: Cредняя выручка пользователей из Москвы не равна выручке пользователей из других регионов
    

</div>

```python
gen_moscow = general.query('city == "Москва"')
gen_not_moscow = general.query('city != "Москва"')
alpha = .05
results = st.ttest_ind(gen_moscow['pay'],gen_not_moscow['pay'])
display(results.pvalue)
if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не отвергаем нулевую гипозезу')
```

<div class="alert alert-info" role="alert">
Получается средняя выручка с пользователей из Москвы особо не отличается от выручки с пользователей других регионов.
</div>
